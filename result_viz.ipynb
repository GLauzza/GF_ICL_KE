```python
import json
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
ids = []
acc = []
gen = []
spe = []

# Interruption by OS !!!! result will not be reliable
# Case 38 is the hardest
for i in range(1, 25):
    summary_path = os.path.join(f"./results/ModelEdit/run_{i:03}", "summary.json")
    if os.path.exists(summary_path):
        with open(summary_path) as f:
            summary = json.load(f)[0]
            if summary["num_cases"] == 50:
                ids.append(i-1)
                acc.append(summary["post_rewrite_success"][0])
                gen.append(summary["post_paraphrase_success"][0])
                spe.append(summary["post_neighborhood_success"][0])

print(ids)

plt.plot(ids, acc, label="Edit Success")
plt.plot(ids, gen, label= "Generalization")
plt.plot(ids, spe, label="Specificity")
plt.hlines(summary["pre_rewrite_success"][0], 0, 24, colors="blue")
plt.hlines(summary["pre_paraphrase_success"][0], 0, 24, colors="orange")
plt.hlines(summary["pre_neighborhood_success"][0], 0, 24, colors="green")
plt.xlabel("edited layer")
plt.ylabel("metric")
plt.legend()
```

```python
acc_mat = []
gen_mat = []
spe_mat = []

for i in range(24):
    acc = []
    gen = []
    spe = []
    for j in range(11):
        summary_path = os.path.join(f"./results/ModelEdit/run_{i*11+j+26:03}", "summary.json")
        if os.path.exists(summary_path):
            with open(summary_path) as f:
                summary = json.load(f)[0]
                acc.append(summary["post_rewrite_diff"][0])
                gen.append(summary["post_paraphrase_diff"][0])
                spe.append(summary["post_neighborhood_diff"][0])
    acc_mat.append(acc)
    gen_mat.append(gen)
    spe_mat.append(spe)
acc_mat = np.array(acc_mat).T
gen_mat = np.array(gen_mat).T
spe_mat = np.array(spe_mat).T

plt.imshow(acc_mat)
plt.colorbar()
plt.xlabel("edited layer")
plt.ylabel("token")
plt.title("Accuracy for single token single layer")
plt.show()
plt.imshow(gen_mat)
plt.colorbar()
plt.xlabel("edited layer")
plt.ylabel("token")
plt.title("Generalization for single token single layer")
plt.show()
plt.imshow(spe_mat)
plt.colorbar()
plt.xlabel("edited layer")
plt.ylabel("token")
plt.title("Specificity for single token single layer")
plt.show()
plt.imshow((spe_mat + gen_mat + acc_mat)/3)
plt.colorbar()
plt.xlabel("edited layer")
plt.ylabel("token")
plt.title("Avg score for multi token single layer")
plt.show()
```
